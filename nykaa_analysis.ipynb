```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pl = pd.DataFrame({
    'Year': ['Mar18','Mar19','Mar20','Mar21','Mar22','Mar23','Mar24','Mar25'],
    'Sales': [27, 102, 204, 146, 188, 275, 313, 420],
    'Expenses': [42, 113, 201, 149, 171, 313, 388, 460],
    'Operating_Profit': [-16, -11, 4, -3, 17, -38, -75, -40],
    'Net_Profit': [-8, 1, 15, 36, 104, 57, 121, 97],
})

print(pl)
```

```python
bs = pd.DataFrame({
    'Year': ['Mar18','Mar19','Mar20','Mar21','Mar22','Mar23','Mar24','Mar25'],
    'Equity_Capital': [14, 14, 15, 15, 47, 285, 286, 286],
    'Reserves': [130, 279, 404, 546, 1503, 1170, 1249, 1374],
    'Borrowings': [8, 30, 40, 46, 51, 55, 87, 86],
    'Total_Assets': [165, 354, 498, 647, 1690, 1836, 1868, 1890],
    'Fixed_Assets': [6, 14, 18, 13, 16, 29, 30, 39],
    'Investments': [22, 169, 52, 70, 379, 461, 641, 935],
})

cf = pd.DataFrame({
    'Year': ['Mar18','Mar19','Mar20','Mar21','Mar22','Mar23','Mar24','Mar25'],
    'Cash_from_Operations': [-62, -56, -90, -84, 72, -0, -31, 143],
    'Cash_from_Investing': [55, -91, 69, -67, -944, -41, -26, -146],
    'Cash_from_Financing': [40, 134, 102, 87, 873, 21, 47, 8],
    'Net_Cash_Flow': [32, -13, 81, -64, 0, -21, -10, 4],
})

print("Balance Sheet:")
print(bs)
print("\nCash Flow:")
print(cf)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Nykaa (FSN E-Commerce) Financial Analysis', fontsize=16, fontweight='bold')

# Chart 1: Revenue Growth
axes[0,0].bar(pl['Year'], pl['Sales'], color='pink')
axes[0,0].set_title('Revenue Growth (Rs. Crores)')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Sales')
axes[0,0].tick_params(axis='x', rotation=45)

# Chart 2: Net Profit
axes[0,1].bar(pl['Year'], pl['Net_Profit'], color=['red' if x < 0 else 'green' for x in pl['Net_Profit']])
axes[0,1].set_title('Net Profit (Rs. Crores)')
axes[0,1].set_xlabel('Year')
axes[0,1].tick_params(axis='x', rotation=45)
axes[0,1].axhline(y=0, color='black', linewidth=0.8)

# Chart 3: Total Assets
axes[1,0].plot(bs['Year'], bs['Total_Assets'], marker='o', color='blue')
axes[1,0].set_title('Total Assets Growth (Rs. Crores)')
axes[1,0].set_xlabel('Year')
axes[1,0].tick_params(axis='x', rotation=45)

# Chart 4: Cash Flow
axes[1,1].plot(cf['Year'], cf['Cash_from_Operations'], marker='o', label='Operations', color='green')
axes[1,1].plot(cf['Year'], cf['Net_Cash_Flow'], marker='s', label='Net Cash', color='red')
axes[1,1].set_title('Cash Flow Trend (Rs. Crores)')
axes[1,1].legend()
axes[1,1].tick_params(axis='x', rotation=45)
axes[1,1].axhline(y=0, color='black', linewidth=0.8)

plt.tight_layout()
plt.show()
```
